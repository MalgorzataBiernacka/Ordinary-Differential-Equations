```python
from numpy import cos, sin, pi, linspace, arange,ones, zeros
```

```python
import matplotlib.pyplot as plt
```

```python
def a(n):
    return (2/(pi*n)**2)*(cos(n*pi)-1)
```

```python
def f(x,N):
    return 1/2*ones(len(x))+sum(a(n)*cos(n*pi*x) for n in arange(1,N,1))
```

```python
x=linspace(-1,3,501)
```

```python
plt.plot(x,f(x,1))
```

```python
plt.plot(x,f(x,2))
```

```python
def y1(x,m,n):
    return sin(m*pi*x)*sin(n*pi*x)

def y2(x):
    return zeros(len(x))

x=linspace(1,3,251)
```

```python
plt.fill_between(x,y1(x,2,1),y2(x),where= y2(x)<=y1(x,2,1),facecolor='lightsteelblue')
plt.fill_between(x,y1(x,2,1),y2(x),where= y1(x,2,1)<=y2(x),facecolor='lightgreen')
```
